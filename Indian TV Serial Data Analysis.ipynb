```python
serial_year = df_serial.sort_values(by = 'Rating', ascending = False)

# Create a bar plot with Seaborn
plt.figure(figsize=(12, 10))
bar_plot = sns.barplot(x = 'TV Serial Name', y = 'Rating', hue = 'Year', data=serial_year)

# Display data labels on each bar for 'Year' column
for p in bar_plot.patches:
    bar_plot.annotate(f'{p.get_height():.1f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha = 'center', va = 'center', xytext=(0, 10), textcoords = 'offset points', fontsize=10)

# Rotate x-axis labels for better readability
plt.xticks(rotation = 90, ha = 'right')

# Adding labels and title
plt.xlabel('TV Serial Name')
plt.ylabel('Rating')
plt.title('TV Serial Ratings by Year')

# Show the plot
plt.tight_layout()
plt.legend(title='Year',bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

```python
# Group by 'Channel Name' and calculate the average rating
channel_ratings = df_serial.groupby('Channel Name')['Rating'].sum().reset_index()

channel_ratings = channel_ratings.sort_values(by = 'Rating', ascending = False)

# Create a bar plot
plt.figure(figsize=(10, 6))

sns.barplot(x = 'Channel Name', y = 'Rating', data=channel_ratings, hue = 'Channel Name')

plt.title('Total Ratings by Channel')

plt.xlabel('Channel Name')

plt.ylabel('Total Rating')

plt.show()
```

```python
# Function to convert duration to minutes
def duration_to_minutes(duration):
    hours, minutes = 0, 0
    
    if 'hr' in duration:
        hours_str, rest = map(str.strip, duration.split('hr'))
        hours = int(hours_str)
    else:
        rest = duration
    
    if 'min' in rest:
        minutes = int(rest.replace('min', '').strip())

    total_minutes = (hours * 60) + minutes
    return total_minutes

# Apply the function to the 'Duration' column
df_serial['Duration (minutes)'] = df_serial['Duration'].apply(duration_to_minutes)

# Display the DataFrame with the new column
df_serial[['TV Serial Name', 'Duration', 'Duration (minutes)']]
```

```python
# Assuming you already have the DataFrame 'df' containing the TV serial data

# Calculate the length of tweets for each TV Serial Name
tv_serials['Tweet Length'] = tv_serials['Tweets'].apply(lambda x: len(str(x)) if pd.notna(x) else 0)

tv_serials = tv_serials.sort_values(by = 'Tweet Length', ascending = False)

# Display the DataFrame with the calculated tweet lengths
tv_serials[['TV Serial Name', 'Tweet Length']]
```

```python
sns.barplot(x = 'Duration (minutes)', y = 'TV Serial Name' , data = tv_serials , hue = 'Duration (minutes)')
plt.xlabel('Duration (minutes)')
plt.ylabel('TV Serial Name')
plt.title('TV Serial Name by Duration (minutes)')
plt.show()
```

```python
# Create a pie chart for genres distribution
genres_counts = tv_serials['Genres'].head(12).str.split(', ').explode().value_counts()
plt.figure(figsize = (8, 8))
plt.pie(genres_counts, labels=genres_counts.index, autopct = '%1.1f%%', startangle=90)
plt.title('Genres Distribution of TV Serials')
plt.show()
```

```python
# Create a scatter plot for Year vs. Ratings
plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'Year', y = 'Rating', data = tv_serials, hue = 'Year', size = 'Duration', sizes = (10, 200))
plt.title('Year vs. Ratings of TV Serials')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.show()
```

```python
# Create a line plot for ratings trend over the years
plt.figure(figsize = (12, 6))
sns.lineplot(x = 'Year', y = 'Rating', data=tv_serials, hue = 'Language', marker = 'o')
plt.title('Ratings Trend Over the Years')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.legend(title = 'Language', loc = 'best')
plt.show()
```

```python
# Create a count plot for channel-wise distribution
plt.figure(figsize = (10, 6))
sns.countplot(y = 'Channel Name', data = tv_serials, order = tv_serials['Channel Name'].value_counts().index)
plt.title('Distribution of TV Serials Across Channels')
plt.xlabel('Number of TV Serials')
plt.ylabel('Channel Name')
plt.show()
```

```python
# Create a box plot for duration distribution by genre
plt.figure(figsize = (12, 8))
sns.boxplot(x = 'Genres', y = 'Duration (minutes)', data = tv_serials)
plt.title('Duration Distribution by Genre')
plt.xlabel('Genres Namw')
plt.ylabel('Duration (minutes)')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```
